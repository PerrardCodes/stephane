```python
def get_mask(im,p):
    [nx,ny] = im.shape
    Nx = int(np.ceil(nx/p))
    Ny = int(np.ceil(ny/p))

    x = np.floor(np.arange(0,Nx,1/p))
    y = np.floor(np.arange(0,Ny,1/p))

    A = np.transpose(np.tile(x,(len(y),1)))
    B = np.transpose(np.tile(np.reshape(y,(len(y),1)),(1,len(x))))

    mask = np.ndarray.astype(A*Ny+B+1,int)
    mask = mask[:nx,:ny]
    #print(Nx*Ny/np.max(mask))
    return mask
```
